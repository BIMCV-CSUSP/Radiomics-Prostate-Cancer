```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_images(image1, image2, image3, types, title = "Comparación de Secuencias MRI"):  
    image_array1 = sitk.GetArrayViewFromImage(image1)
    image_array2 = sitk.GetArrayViewFromImage(image2)
    image_array3 = sitk.GetArrayViewFromImage(image3)

    slice_index1 = image_array1.shape[0] // 2
    slice_index2 = image_array2.shape[0] // 2
    slice_index3 = image_array3.shape[0] // 2

    slice1 = image_array1[slice_index1]
    slice2 = image_array2[slice_index2]
    slice3 = image_array3[slice_index3]

    fig, axs = plt.subplots(1, 3, figsize=(18, 6))  
    
    axs[0].imshow(slice1, cmap="gray")
    axs[0].set_title(f"Secuencia {types[0]}", fontsize=16, pad=20) 
    axs[0].axis("off")

    axs[1].imshow(slice2, cmap="gray")
    axs[1].set_title(f"Secuencia {types[1]}", fontsize=16, pad=20)
    axs[1].axis("off")

    axs[2].imshow(slice3, cmap="gray")
    axs[2].set_title(f"Secuencia {types[2]}", fontsize=16, pad=20)
    axs[2].axis("off")

    plt.suptitle("Comparación de Secuencias MRI", fontsize=20, y=1.02)
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()
```

```python
def load_images(patient_id, study_id):
    image_path_t2w = f"../../../data/images/{patient_id}/{patient_id}_{study_id}_t2w.mha"
    image_t2w = sitk.ReadImage(image_path_t2w)

    image_path_dwi = f"../../../data/images/{patient_id}/{patient_id}_{study_id}_hbv.mha"
    image_dwi = sitk.ReadImage(image_path_dwi)

    image_path_adc = f"../../../data/images/{patient_id}/{patient_id}_{study_id}_adc.mha"
    image_adc = sitk.ReadImage(image_path_adc)
    
    return image_t2w, image_dwi, image_adc
```

```python
mri_sec = ['T2W', 'DWI', 'ADC']
```

# ISUP = 0

```python
image_t2w, image_dwi, image_adc = load_images("10000", "1000000")
plot_images(image_t2w, image_dwi, image_adc, mri_sec)
```

# ISUP = 5

```python
image_t2w, image_dwi, image_adc = load_images("10029", "1000029")
plot_images(image_t2w, image_dwi, image_adc, mri_sec)
```
