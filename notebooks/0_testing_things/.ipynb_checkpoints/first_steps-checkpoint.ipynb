```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
image_path = "../../../data/images/10000/10000_1000000_t2w.mha"
image = sitk.ReadImage(image_path)

image_array = sitk.GetArrayViewFromImage(image)

slice_index = image_array.shape[0] // 2
selected_slice = image_array[slice_index]

# Visualizar la capa seleccionada
plt.imshow(selected_slice, cmap="gray")
plt.title("Visualización de imagen T2W (slice)")
plt.axis("off")
plt.show()
```
