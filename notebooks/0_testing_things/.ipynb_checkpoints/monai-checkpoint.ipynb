```python
import monai
```

```python
patient_id, study_id = "11002", "1001021"

image_dict = {'t2' : f"../../../data/images/{patient_id}/{patient_id}_{study_id}_t2w.mha",
              'dwi' : f"../../../data/images/{patient_id}/{patient_id}_{study_id}_hbv.mha",
              'adc' : f"../../../data/images/{patient_id}/{patient_id}_{study_id}_adc.mha",
              'label' : f"../../../data/labels/csPCa_lesion_delineations/human_expert/resampled/{patient_id}_{study_id}.nii.gz"}

img_columns = ['t2','dwi','adc', 'label']

transforms = monai.transforms.Compose([
                monai.transforms.LoadImaged(keys=img_columns, image_only=False, ensure_channel_first = True),
                monai.transforms.ResampleToMatchd(
                    keys=["adc", "dwi"],
                    key_dst="t2",
                    mode=("bilinear", "bilinear"),
                ), # Resample images to t2 dimension
#                 monai.transforms.Resized(
#                     keys=['t2','dwi_0','adc'],
#                     spatial_size=eval(input_shape),
#                     mode=("trilinear", "trilinear", "trilinear"),
#                 ),
                monai.transforms.ScaleIntensityd(keys=['t2','dwi','adc'], minv=0.0, maxv=1.0, allow_missing_keys=True),
                monai.transforms.ConcatItemsd(keys=['t2','dwi','adc'], name="image", dim=0)
                ])
```

```python
processed_images = transforms(image_dict)
```

```python
t2 = processed_images["image"]
```

```python
monai.visualize.matshow3d(volume=t2, frame_dim=-1, channel_dim=0, every_n=1, margin=10, show=True)
```

```python
t2 = processed_images["t2"]
```

```python
blended = monai.visualize.utils.blend_images(t2,
                                            processed_images["label"],
                                            alpha=0.5,
                                            transparent_background=True)
```

```python
import numpy as np

blended_rotated = np.rot90(blended, k=1, axes=(1, 2))
```

```python
monai.visualize.matshow3d(volume=blended, frame_dim=-1, channel_dim=0, every_n=1, margin=10, show=True)
```

```python
monai.visualize.matshow3d(volume=blended_rotated, frame_dim=-1, channel_dim=0, every_n=1, margin=10, show=True)
```
