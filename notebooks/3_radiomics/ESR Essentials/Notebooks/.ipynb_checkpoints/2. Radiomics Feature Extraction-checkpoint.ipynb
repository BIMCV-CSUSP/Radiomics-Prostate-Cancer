# Datos

```python
df = pd.read_csv("../../../../../data/data.csv")
```

```python
df_nuevo = df[['patient_id', 'study_id', 't2w_path', 'csPCa_lesion_delineation_path']].copy()
df_nuevo['ID'] = df_nuevo['patient_id'].astype(str) + '_' + df_nuevo['study_id'].astype(str)
df_nuevo = df_nuevo[['ID', 't2w_path', 'csPCa_lesion_delineation_path']]
df_nuevo = df_nuevo.rename(columns={
    't2w_path': 'Image',
    'csPCa_lesion_delineation_path': 'Mask'
})
df_nuevo['Image'] = '../../../../../' + df_nuevo['Image'].astype(str)
df_nuevo['Mask']  = '../../../../../' + df_nuevo['Mask'].astype(str)

df_nuevo_sample = df_nuevo.sample(50, random_state=42)
```

```python
df_nuevo_sample.to_csv("input_t2w.csv", index=False)
```
