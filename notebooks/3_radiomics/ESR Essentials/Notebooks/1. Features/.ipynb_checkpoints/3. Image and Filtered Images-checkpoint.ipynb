# Librerias

```python
import SimpleITK as sitk
import matplotlib.pyplot as plt

from skimage.feature import local_binary_pattern
import pywt
```

# Imágenes

```python
pre_path = "../../../../../data/images/"
t2_path = pre_path + "10029/10029_1000029_t2w.mha"
adc_path = pre_path + "10029/10029_1000029_adc.mha"
dwi_path = pre_path + "10029/10029_1000029_hbv.mha"
```

```python
t2_image = sitk.ReadImage(t2_path)
adc_image = sitk.ReadImage(adc_path)
dwi_image = sitk.ReadImage(dwi_path)
```

# Visualización

```python
def visualize_image(image):
    fig, axes = plt.subplots (nrows = 2, ncols = 3, figsize = (10, 10))
    rect = fig.patch
    rect.set_facecolor('black')
    
    axes[0,0].imshow(sitk.GetArrayFromImage(image)[3,:,:], cmap='gray')
    axes[0,1].imshow(sitk.GetArrayFromImage(image)[5,:,:], cmap='gray')
    axes[0,2].imshow(sitk.GetArrayFromImage(image)[8,:,:], cmap='gray')
    axes[1,0].imshow(sitk.GetArrayFromImage(image)[10,:,:], cmap='gray')
    axes[1,1].imshow(sitk.GetArrayFromImage(image)[12,:,:], cmap='gray')
    axes[1,2].imshow(sitk.GetArrayFromImage(image)[14,:,:], cmap='gray')
    
    for ax in axes.flat:
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    fig.tight_layout(h_pad=-15, w_pad=1)
```

```python
visualize_image(t2_image)
```

# Create a copy of the prostate image in float 64-bits

```python
print(t2_image.GetPixelIDTypeAsString())
print(adc_image.GetPixelIDTypeAsString())
print(dwi_image.GetPixelIDTypeAsString())
```

```python
t2_image_float64 = sitk.Cast(t2_image, sitk.sitkFloat64)
adc_image_float64 = sitk.Cast(adc_image, sitk.sitkFloat64)
dwi_image_float64 = sitk.Cast(dwi_image, sitk.sitkFloat64)
```

# Filtros

```python
def process_image(image_float64, lbp_radius=1, lbp_samples=8, lbp_method='uniform', lbp_axis=0):
    # Aplicar filtros con SimpleITK
    image_log_sigma_1 = sitk.LaplacianRecursiveGaussian(image_float64, sigma=1)
    image_log_sigma_5 = sitk.LaplacianRecursiveGaussian(image_float64, sigma=5)
    image_exponential = sitk.Exp(sitk.Normalize(image_float64))
    image_logarithm = sitk.Log(image_float64)
    image_square = sitk.Square(image_float64)
    image_sqrt = sitk.Sqrt(image_float64)
    image_gradient = sitk.GradientMagnitude(image_float64, useImageSpacing=True)

    # Aplicar Local Binary Pattern slice a slice
    im_arr = sitk.GetArrayFromImage(image_float64)
    # Se intercambia el eje si se desea aplicar LBP sobre otro eje (por defecto eje 0)
    im_arr = im_arr.swapaxes(0, lbp_axis)
    for idx in range(im_arr.shape[0]):
        im_arr[idx, ...] = local_binary_pattern(im_arr[idx, ...],
                                                P=lbp_samples,
                                                R=lbp_radius,
                                                method=lbp_method)
    im_arr = im_arr.swapaxes(0, lbp_axis)
    image_lbp2d = sitk.GetImageFromArray(im_arr)
    image_lbp2d.CopyInformation(image_float64)

    # Extraer una slice para la transformada wavelet (por ejemplo, la slice 8)
    im_arr_full = sitk.GetArrayFromImage(image_float64)
    if im_arr_full.shape[0] > 8:
        original_slice = im_arr_full[8, :, :]
    else:
        # Si la imagen tiene menos de 9 slices, se toma la slice central
        center = im_arr_full.shape[0] // 2
        original_slice = im_arr_full[center, :, :]
        
    # Transformada wavelet 2D
    coeffs2 = pywt.dwt2(original_slice, 'bior1.3')
    LL, (LH, HL, HH) = coeffs2

    # Empaquetar resultados en un diccionario
    results = {
        'original': image_float64,
        'log_sigma_1': image_log_sigma_1,
        'log_sigma_5': image_log_sigma_5,
        'exponential': image_exponential,
        'logarithm': image_logarithm,
        'square': image_square,
        'sqrt': image_sqrt,
        'gradient': image_gradient,
        'lbp2d': image_lbp2d,
        'wavelet': {
            'LL': LL,
            'LH': LH,
            'HL': HL,
            'HH': HH
        }
    }
    return results
```

```python
def plot_processed_images(results, slice_idx=8):
    """
    Muestra en una figura algunos resultados del procesamiento.
    
    Parámetros:
      results: dict con las imágenes procesadas (resultado de process_image).
      slice_idx: índice de la slice a visualizar (para imágenes 3D).
    """
    # Extraer imágenes de SimpleITK a arrays para la slice indicada
    original_arr = sitk.GetArrayFromImage(results['original'])[slice_idx, :, :]
    log_sigma_1_arr = sitk.GetArrayFromImage(results['log_sigma_1'])[slice_idx, :, :]
    log_sigma_5_arr = sitk.GetArrayFromImage(results['log_sigma_5'])[slice_idx, :, :]
    exponential_arr = sitk.GetArrayFromImage(results['exponential'])[slice_idx, :, :]
    logarithm_arr = sitk.GetArrayFromImage(results['logarithm'])[slice_idx, :, :]
    square_arr = sitk.GetArrayFromImage(results['square'])[slice_idx, :, :]
    sqrt_arr = sitk.GetArrayFromImage(results['sqrt'])[slice_idx, :, :]
    gradient_arr = sitk.GetArrayFromImage(results['gradient'])[slice_idx, :, :]
    lbp2d_arr = sitk.GetArrayFromImage(results['lbp2d'])[slice_idx, :, :]
    
    # Extraer coeficientes wavelet
    wavelet = results['wavelet']
    LL = wavelet['LL']
    LH = wavelet['LH']
    HL = wavelet['HL']
    HH = wavelet['HH']
    
    titles_wavelet = ['Wv-LL: Aproximación', 'Wv-LH: Detalle horizontal',
                      'Wv-HL: Detalle vertical', 'Wv-HH: Detalle diagonal']
    
    # Configurar figura y ejes
    fig, axes = plt.subplots(nrows=2, ncols=7, figsize=(20, 20))
    fig.patch.set_facecolor('black')
    
    # Fila superior: imágenes transformadas
    axes[0,0].imshow(original_arr, cmap='gray')
    axes[0,1].imshow(log_sigma_1_arr, cmap='gray')
    axes[0,2].imshow(log_sigma_5_arr, cmap='gray')
    axes[0,3].imshow(exponential_arr, cmap='gray')
    axes[0,4].imshow(logarithm_arr, cmap='gray')
    axes[0,5].imshow(square_arr, cmap='gray')
    axes[0,6].imshow(sqrt_arr, cmap='gray')
    
    # Fila inferior: gradiente, LBP y wavelet
    axes[1,0].imshow(gradient_arr, cmap='gray')
    axes[1,1].imshow(lbp2d_arr, cmap='gray')
    axes[1,2].imshow(LL, cmap='gray')
    axes[1,3].imshow(LH, cmap='gray')
    axes[1,4].imshow(HL, cmap='gray')
    axes[1,5].imshow(HH, cmap='gray')
    
    # Eliminar eje sobrante (último subplot de la fila inferior)
    fig.delaxes(axes[1][6])
    
    # Títulos de cada subplot
    axes[0,0].set_title('Original', color='w')
    axes[0,1].set_title('LoG Sigma=1', color='w')
    axes[0,2].set_title('LoG Sigma=5', color='w')
    axes[0,3].set_title('Exponential', color='w')
    axes[0,4].set_title('Logarithm', color='w')
    axes[0,5].set_title('Square', color='w')
    axes[0,6].set_title('Square-Root', color='w')
    
    axes[1,0].set_title('Gradient', color='w')
    axes[1,1].set_title('LBP2D', color='w')
    axes[1,2].set_title(titles_wavelet[0], color='w')
    axes[1,3].set_title(titles_wavelet[1], color='w')
    axes[1,4].set_title(titles_wavelet[2], color='w')
    axes[1,5].set_title(titles_wavelet[3], color='w')
    
    # Ocultar ejes
    for ax in axes.flat:
        ax.axis('off')
    fig.tight_layout(h_pad=-75, w_pad=1)
    plt.show()
```

```python
t2_results = process_image(t2_image_float64)
plot_processed_images(t2_results, slice_idx=8)
```

```python
adc_results = process_image(adc_image_float64)
plot_processed_images(adc_results, slice_idx=8)
```

```python
dwi_results = process_image(dwi_image_float64)
plot_processed_images(dwi_results, slice_idx=8)
```
